```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Define the text variable with a string containing the words for the word cloud.
text = "This is an example string for generating a word cloud. You can replace this with your desired text."

plt.figure(figsize=(10,10))
wordcloud = WordCloud(background_color='white').generate(text)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from PIL import Image

# Load the Python logo mask (You need a black-and-white Python logo image)
mask = np.array(Image.open("/content/Python.png"))

# Define the text for the word cloud
text = """Python Programming Language OpenAI Machine Learning Data Science
          Artificial Intelligence Deep Learning Pandas NumPy Matplotlib
          Scikit-learn TensorFlow PyTorch Flask Django FastAPI"""

# Generate the word cloud
wordcloud = WordCloud(
    width=800,
    height=800,
    background_color="white",
    mask=mask,
    contour_width=2,
    contour_color="blue",
    colormap="cool",
).generate(text)

# Display the word cloud
plt.figure(figsize=(10, 10))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
